## Risk Stratify Patients

```python


import pandas as pd
```

```python
df = pd.read_excel('w5e6_patient_risk_score_cost.xlsx')
```

```python
df = df.set_index('id')
```

```python
df.info(verbose=True)
```

```python
# Stratify by all risk scores (sum)
df['risk_score_total'] = df['comm_health_needs_score'] + df['risk_score_morbidity'] + df['risk_score_cost']
```

```python
# Stratficiation Options
# DECISION: How many strata do we want (minimum of 2)? (And, do we want the strata to be balanced or unbalanced?)

# Split into classes by quantile, based on numof_strata (low numbers are low risk; high nubmers are high risk)
# numof_strata = 3
# df['risk_strata'] = pd.qcut(df['risk_score_total'], q=numof_strata, labels=False, precision=0)

# (or, unequally sized strata can also be created)
# 5 strata(0-5) then binned where: where 0,1,2 is low risk (0), 3,4 is moderate risk (1), 5 is high risk (2)
import numpy as np
df['risk_strata'] = pd.qcut(df['risk_score_total'], q=6, labels=False, precision=0)
df['risk_strata'] = pd.cut(df['risk_strata'], bins=[-np.inf,2,4,5], labels=[0, 1, 2])
df['risk_strata'] = df['risk_strata'].map({0:'low', 1:'med', 2:'high'})
```

```python
# Output to file
df.to_excel('w6e1_patient_risk_strata.xlsx')
```
